```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
t = np.arange(0, 12, 0.01)
y = np.sin(t)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Sinewave Demo')
plt.plot(t, y)
```

```python
plt.figure(figsize=(10,8))
t = [0,1,2,3,4,5,6]
y = [10, 20, 30, 10, 5, 20, 30]

plt.plot(t, y, color='red', linestyle='dashed', marker="o", markerfacecolor="blue", markersize=14)
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
print(tips.head())
print(tips.info())
```

```python
scatter_plot = plt.figure()
axes = scatter_plot.add_subplot(1,1,1)
axes.scatter(tips['total_bill'], tips['tip'])
axes.set_title('Scatterplot of Total Bill vs Tip')
axes.set_xlabel('Total Bill')
axes.set_ylabel('Tip')
```

```python
names = ['group_a','group_b','group_c']
values = [1, 10, 100]

plt.figure(figsize=(12, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('카테고리별 차트')
```

```python
import platform
from matplotlib import font_manager, rc

# Matplotlib 은 기본적으로 음수 표시를 일반 - 문자가 아니라 유니코드 마이너스 기호(-)를 사용해서 한글 폰트에서는 깨짐. 그래서 사용하지 않도록 설정.
plt.rcParams['axes.unicode.minus'] = False 

if platform.system() == "Darwin":
    rc('font', family='AppleGothic')
elif platform.system() == "Windows":
    path = 'C:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system')
```
