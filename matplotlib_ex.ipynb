```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
t = np.arange(0, 12, 0.01)
y = np.sin(t)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Sinewave Demo')
plt.plot(t, y)
```

```python
# df.loc[[행],[열]] 이나 df.iloc[[행],[열]]과 같이 지정하면 됨.
# 다음은 모든 행(:)의 데이터에 대해 year, pop 열을 추출하는 방법임.
subset = df.loc[:, ["year","pop"]]
```

```python
subset.head()
```

```python
# 기본적인 통계는 아래와 같이 작성할 수 있음.
# 연도별로 lifeExp열의 평균을 계산함. 데이터를 year 열을 기준으로 그룹화하고, lifeExp 열의 평균을 구하면 됨
df.groupby("year")["lifeExp"].mean()
```

```python
#lifeExp, gdpPercap 열의 평균값을 연도, 지역별로 그룹화하여 한번에 계산할 수있음.
#그룹을 만드는 기준이 아닌 경우, 아래와 같이 ["year","continent"]를 묶어서 그룹에 대한 기준을 지정하고, 두 번째 리스트로 ["lifeExp","gdpPercap"]를 지정해서 mean() 메서드로 평균을 구함.
multi_group = df.groupby(["year","continent"])[["lifeExp","gdpPercap"]].mean()
```

```python
multi_group.head()
```

```python
# 아래와 같이 연도별로 그룹을 만들고, 기대 수명의 평균값을 구해서 차트를 생성할 수 있음.
# groupby()에서 "year"를 지정해서 기준을 정하고, "lifeExp"컬럼을 지정해서 mean() 평균값을 구하면 됨.
# 5년에 한번씩 설문을 했기 때문에 1952년, 1957년...과 같이 5년마다 집계된 데이터가 연도별로 출력됨.

%matplotlib inline
import matplotlib.pyplot as plt

global_yearly_lifeExp = df.groupby("year")["lifeExp"].mean()
global_yearly_lifeExp
```

```python
# 마지막으로 라인 그래프 형태로 그림.
global_yearly_lifeExp.plot()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 타이타닉 데이터셋을 인터넷으로 불러오기(seaborn 내장 데이터 사용)
titanic = sns.load_dataset('titanic')

# 성별 생존자 수 집계
survival_by_sex = titanic.groupby(['sex', 'survived']).size().unstack() # size(): 각 그룹의 행 개수 계산, unstack() : 행 인덱스를 열로 펼치는 작업.

# 그래프 스타일 설정(선택사항)
sns.set(style="whitegrid")

# 생존자 비교 바 차트
survival_by_sex.plot(kind='bar', stacked=False) # stacked=False : 막대를 겹쳐 쌓지 않고 옆으로 나란히 표시.
plt.title('Survival Count by Gender')
plt.xlabel("Sex")
plt.ylabel("Number of People")
plt.legend(['Did not survive', 'Survived'], title='Survival')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
